```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_details.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
param_grid={'n_estimators':[50,100],
               'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
            'max_features':['sqrt','log2',None]}
```

```python
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
regressor=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(regressor)
```

```python
table
```

```python
grid.best_params_
```

```python
import pickle
Filename="Final RF - GridSearchCV.sav"
pickle.dump(grid,open(Filename,'wb'))
```

```python
load=pickle.load(open('Final RF - GridSearchCV.sav','rb'))
result=load.predict([[24,17,0,0,1]])
```
